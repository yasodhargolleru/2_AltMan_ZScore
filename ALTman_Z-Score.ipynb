```python
def z_altman0(tick):

    current_assets = tick.balance_sheet.loc['Total Assets'][0]
    current_liabilities = tick.balance_sheet.loc['Total Liabilities Net Minority Interest'][0]
    ebit = tick.financials.loc['EBIT'][0]
    sales = tick.financials.loc['Total Revenue'][0]
    retained_earnings = tick.balance_sheet.loc['Retained Earnings'][0]
    market_cap = tick.info['marketCap']

    working_capital = current_assets - current_liabilities

    working_capital_to_assets = working_capital / current_assets
    retained_earnings_to_assets = retained_earnings / current_assets
    ebit_to_assets = ebit / current_assets
    market_cap_to_liabilities = market_cap / current_liabilities
    sales_to_assets = sales / current_assets

    z_score = (
        (1.2 * working_capital_to_assets) +
        (1.4 * retained_earnings_to_assets) +
        (3.3 * ebit_to_assets) +
        (0.6 * market_cap_to_liabilities) +
        (1.0 * sales_to_assets)
    )

    return z_score

def z_altman1(tick):

    current_assets = tick.balance_sheet.loc['Total Assets'][1]
    current_liabilities = tick.balance_sheet.loc['Total Liabilities Net Minority Interest'][1]
    ebit = tick.financials.loc['EBIT'][1]
    sales = tick.financials.loc['Total Revenue'][1]
    retained_earnings = tick.balance_sheet.loc['Retained Earnings'][1]
    market_cap = tick.info['marketCap']

    working_capital = current_assets - current_liabilities

    working_capital_to_assets = working_capital / current_assets
    retained_earnings_to_assets = retained_earnings / current_assets
    ebit_to_assets = ebit / current_assets
    market_cap_to_liabilities = market_cap / current_liabilities
    sales_to_assets = sales / current_assets

    z_score = (
        (1.2 * working_capital_to_assets) +
        (1.4 * retained_earnings_to_assets) +
        (3.3 * ebit_to_assets) +
        (0.6 * market_cap_to_liabilities) +
        (1.0 * sales_to_assets)
    )

    return z_score

def z_altman2(tick):


    current_assets = tick.balance_sheet.loc['Total Assets'][2]
    current_liabilities = tick.balance_sheet.loc['Total Liabilities Net Minority Interest'][2]
    ebit = tick.financials.loc['EBIT'][2]
    sales = tick.financials.loc['Total Revenue'][2]
    retained_earnings = tick.balance_sheet.loc['Retained Earnings'][2]
    market_cap = tick.info['marketCap']

    working_capital = current_assets - current_liabilities

    working_capital_to_assets = working_capital / current_assets
    retained_earnings_to_assets = retained_earnings / current_assets
    ebit_to_assets = ebit / current_assets
    market_cap_to_liabilities = market_cap / current_liabilities
    sales_to_assets = sales / current_assets

    z_score = (
        (1.2 * working_capital_to_assets) +
        (1.4 * retained_earnings_to_assets) +
        (3.3 * ebit_to_assets) +
        (0.6 * market_cap_to_liabilities) +
        (1.0 * sales_to_assets)
    )

    return z_score

def z_altman3(tick):


    current_assets = tick.balance_sheet.loc['Total Assets'][3]
    current_liabilities = tick.balance_sheet.loc['Total Liabilities Net Minority Interest'][3]
    ebit = tick.financials.loc['EBIT'][3]
    sales = tick.financials.loc['Total Revenue'][2]
    retained_earnings = tick.balance_sheet.loc['Retained Earnings'][3]
    market_cap = tick.info['marketCap']

    working_capital = current_assets - current_liabilities

    working_capital_to_assets = working_capital / current_assets
    retained_earnings_to_assets = retained_earnings / current_assets
    ebit_to_assets = ebit / current_assets
    market_cap_to_liabilities = market_cap / current_liabilities
    sales_to_assets = sales / current_assets

    z_score = (
        (1.2 * working_capital_to_assets) +
        (1.4 * retained_earnings_to_assets) +
        (3.3 * ebit_to_assets) +
        (0.6 * market_cap_to_liabilities) +
        (1.0 * sales_to_assets)
    )

    return z_score
```

```python

def z_altman_multi_0(ticks):


    scores_dict = dict()

    # Loop over ticks
    for stock in ticks.tickers:

        # Get numbers
        current_assets = ticks.tickers[stock].balance_sheet.loc['Total Assets'][0]
        current_liabilities = ticks.tickers[stock].balance_sheet.loc['Total Liabilities Net Minority Interest'][0]
        ebit = ticks.tickers[stock].financials.loc['EBIT'][0]
        sales = ticks.tickers[stock].financials.loc['Total Revenue'][0]
        retained_earnings = ticks.tickers[stock].balance_sheet.loc['Retained Earnings'][0]
        market_cap = ticks.tickers[stock].info['marketCap']

        # Calculate working capital
        working_capital = current_assets - current_liabilities

        # Calculate ratios
        working_capital_to_assets = working_capital / current_assets
        retained_earnings_to_assets = retained_earnings / current_assets
        ebit_to_assets = ebit / current_assets
        market_cap_to_liabilities = market_cap / current_liabilities
        sales_to_assets = sales / current_assets

        # Multiply by weights
        z_score = (
            (1.2 * working_capital_to_assets) +
            (1.4 * retained_earnings_to_assets) +
            (3.3 * ebit_to_assets) +
            (0.6 * market_cap_to_liabilities) +
            (1.0 * sales_to_assets)
        )

        # Add score to dict
        scores_dict[stock] = z_score

    return scores_dict

def z_altman_multi_1(ticks):


    scores_dict = dict()

    # Loop over ticks
    for stock in ticks.tickers:

        # Get numbers
        current_assets = ticks.tickers[stock].balance_sheet.loc['Total Assets'][1]
        current_liabilities = ticks.tickers[stock].balance_sheet.loc['Total Liabilities Net Minority Interest'][1]
        ebit = ticks.tickers[stock].financials.loc['EBIT'][1]
        sales = ticks.tickers[stock].financials.loc['Total Revenue'][1]
        retained_earnings = ticks.tickers[stock].balance_sheet.loc['Retained Earnings'][1]
        market_cap = ticks.tickers[stock].info['marketCap']

        # Calculate working capital
        working_capital = current_assets - current_liabilities

        # Calculate ratios
        working_capital_to_assets = working_capital / current_assets
        retained_earnings_to_assets = retained_earnings / current_assets
        ebit_to_assets = ebit / current_assets
        market_cap_to_liabilities = market_cap / current_liabilities
        sales_to_assets = sales / current_assets

        # Multiply by weights
        z_score = (
            (1.2 * working_capital_to_assets) +
            (1.4 * retained_earnings_to_assets) +
            (3.3 * ebit_to_assets) +
            (0.6 * market_cap_to_liabilities) +
            (1.0 * sales_to_assets)
        )

        # Add score to dict
        scores_dict[stock] = z_score

    return scores_dict

def z_altman_multi_2(ticks):


    scores_dict = dict()

    # Loop over ticks
    for stock in ticks.tickers:

        # Get numbers
        current_assets = ticks.tickers[stock].balance_sheet.loc['Total Assets'][2]
        current_liabilities = ticks.tickers[stock].balance_sheet.loc['Total Liabilities Net Minority Interest'][2]
        ebit = ticks.tickers[stock].financials.loc['EBIT'][2]
        sales = ticks.tickers[stock].financials.loc['Total Revenue'][2]
        retained_earnings = ticks.tickers[stock].balance_sheet.loc['Retained Earnings'][2]
        market_cap = ticks.tickers[stock].info['marketCap']

        # Calculate working capital
        working_capital = current_assets - current_liabilities

        # Calculate ratios
        working_capital_to_assets = working_capital / current_assets
        retained_earnings_to_assets = retained_earnings / current_assets
        ebit_to_assets = ebit / current_assets
        market_cap_to_liabilities = market_cap / current_liabilities
        sales_to_assets = sales / current_assets

        # Multiply by weights
        z_score = (
            (1.2 * working_capital_to_assets) +
            (1.4 * retained_earnings_to_assets) +
            (3.3 * ebit_to_assets) +
            (0.6 * market_cap_to_liabilities) +
            (1.0 * sales_to_assets)
        )

        # Add score to dict
        scores_dict[stock] = z_score

    return scores_dict

def z_altman_multi_3(ticks):


    scores_dict = dict()

    # Loop over ticks
    for stock in ticks.tickers:

        # Get numbers
        current_assets = ticks.tickers[stock].balance_sheet.loc['Total Assets'][3]
        current_liabilities = ticks.tickers[stock].balance_sheet.loc['Total Liabilities Net Minority Interest'][3]
        ebit = ticks.tickers[stock].financials.loc['EBIT'][3]
        sales = ticks.tickers[stock].financials.loc['Total Revenue'][3]
        retained_earnings = ticks.tickers[stock].balance_sheet.loc['Retained Earnings'][3]
        market_cap = ticks.tickers[stock].info['marketCap']

        # Calculate working capital
        working_capital = current_assets - current_liabilities

        # Calculate ratios
        working_capital_to_assets = working_capital / current_assets
        retained_earnings_to_assets = retained_earnings / current_assets
        ebit_to_assets = ebit / current_assets
        market_cap_to_liabilities = market_cap / current_liabilities
        sales_to_assets = sales / current_assets

        # Multiply by weights
        z_score = (
            (1.2 * working_capital_to_assets) +
            (1.4 * retained_earnings_to_assets) +
            (3.3 * ebit_to_assets) +
            (0.6 * market_cap_to_liabilities) +
            (1.0 * sales_to_assets)
        )

        # Add score to dict
        scores_dict[stock] = z_score

    return scores_dict


```
